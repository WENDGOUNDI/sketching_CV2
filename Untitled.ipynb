```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as matim
import ipywidgets as widgets
```

```python
def adjust_gamma(image, gamma=1):
    invGamma =1.0 / gamma
    table = np.array([((i/255)**invGamma)*255 for i in np.arange(0,256)])
    lut_img = cv2.LUT(image.astype(np.uint8), table.astype(np.uint8))
    return lut_img
```

```python
path = "image.jpg"
img = matim.imread(path)
```

```python
#pencil sketsch
ksize = 21
sigmax = 1
gamma = .1
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_blur = cv2.GaussianBlur(gray, (ksize,ksize), sigmax)  #ksize=3 to 25 and sigma=1 to 25
gray_blur_divide = cv2.divide(gray, gray_blur, scale=256)
pencil_sketch = adjust_gamma(gray_blur_divide, gamma=gamma)

#Display output

plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
plt.imshow(gray, cmap='gray')
plt.subplot(2,2,2)
plt.imshow(gray_blur, cmap='gray')
plt.subplot(2,2,3)
plt.imshow(gray_blur_divide, cmap='gray')
plt.subplot(2,2,4)
plt.imshow(pencil_sketch, cmap='gray')
```

```python
@widgets.interact_manual(ksize=(3,25,2),sigmaX=(1,15,2), gamma=(0,1,.1))

def pencil_art_image(ksize=21, sigmaX=9, gamma=.1):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_blur = cv2.GaussianBlur(gray, (ksize,ksize), sigmax)  #ksize=3 to 25 and sigma=1 to 25
    gray_blur_divide = cv2.divide(gray, gray_blur, scale=256)
    pencil_sketch = adjust_gamma(gray_blur_divide, gamma=gamma)

        #Display the output
    plt.figure(figsize=(15,8))
    plt.subplot(1,2,1)
    plt.imshow(img)
    plt.yticks([]),plt.xticks([])
    plt.title("Original image")
    plt.subplot(1,2,2)
    plt.imshow(pencil_sketch)
    plt.title("Pencil Sketch")
    plt.yticks([]),plt.xticks([])
    plt.show()
```
