```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as matim
import ipywidgets as widgets
```

```python
img = matim.imread("image.jpg")
plt.imshow(img)
```

```python
#Applying edge preserving filter
# The cv2.edgePreservingFilter() takes two important argument such sigma_s ranging from 0 to 200 and
# sigma_r ranging from 0 to 1 

@widgets.interact_manual(s=(0,200,1),r=(0,1,.1))
def edgePreserve(s=50, r=0.5):
    edgeImg = cv2.edgePreservingFilter(img,sigma_s=s, sigma_r=r)
    #Display the output
    plt.figure(figsize=(15,8))
    plt.subplot(1,2,1)
    plt.imshow(img)
    plt.subplot(1,2,2)
    plt.imshow(edgeImg)
    plt.show()
```

```python
# Detail Enhancing Filter
@widgets.interact_manual(s=(0,200,1),r=(0,1,.1))
def edgePreserve(s=50, r=0.5):
    detialImg = cv2.detailEnhance(img,sigma_s=s, sigma_r=r)
    #Display the output
    plt.figure(figsize=(15,8))
    plt.subplot(1,2,1)
    plt.imshow(img)
    plt.yticks([]),plt.xticks([])
    plt.title("Original image")
    plt.subplot(1,2,2)
    plt.imshow(detialImg)
    plt.title("Detail image")
    plt.yticks([]),plt.xticks([])
    plt.show()
```
